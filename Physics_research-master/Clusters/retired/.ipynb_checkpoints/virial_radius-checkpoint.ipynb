```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
coords_dir = '/storage/cluster/data/'
dens_dir = '/storage/filament/data/box01/256den18a+1024+1'
```

```python
coords = np.genfromtxt(coords_dir + 'coords.txt')
dens = np.genfromtxt(dens_dir)
dens = dens.reshape([256,256,256])
```

```python
data_num =[]
data_coordx =[]
data_coordy =[]
data_coordz =[]
data_radius =[]
data_log_mass = []
data_radius_grid = []
for nn,(ix,iy,iz) in enumerate(coords):
    dens_mean = 0

    for r in range(0,10):
#         try:
#             dens_mean = np.mean(dens[int(ix)-r : int(ix)+r+1 , int(iy)-r : int(iy)+r+1 , int(iz)-r : int(iz)+r+1])
#         except:
#             continue
        dens_cluster = dens[int(ix)-r : int(ix)+r+1 , int(iy)-r : int(iy)+r+1 , int(iz)-r : int(iz)+r+1]
        if dens_cluster.shape[0] == dens_cluster.shape[1] == dens_cluster.shape[2]:
            dens_mean = np.mean(dens_cluster)
        else:
            continue

        if dens_mean*(0.28/0.044) <= 200:
            data_num.append(int(nn))
            data_coordx.append(int(ix*4))
            data_coordy.append(int(iy*4))
            data_coordz.append(int(iz*4))
            data_radius.append(r*0.39)
            data_radius_grid.append(r)
            m = 3*(np.log10(r*0.39*0.7*3.0856*10**24)) + np.log10(dens_mean*(1.879*10**-29)*(1.428**2)*(0.28)) + np.log10(4*np.pi/3)
            #print(3*(np.log10(r*0.39* ((0.7)*3.0856*10**24))) , np.log10(dens_mean*(1.879*10**-29)*(1.428**2)*(0.28)) )
            data_log_mass.append(m)
            #print(dens_mean)
            
            break
    
```

```python
df = pd.DataFrame({'number':data_num,'x':data_coordx,'y':data_coordy,'z':data_coordz,'virial_radius (Mpc)':data_radius, \
                  'data_mass (log)': data_log_mass})
df.to_excel('/storage/cluster/data/test.xlsx')
```

```python
xray_dir = '/storage/filament/data/box01/256den18g+1024+1'
temp_dir = '/storage/filament/data/box01/256den18f+1024+1'
```

```python
xray = np.genfromtxt(xray_dir)
xray = xray.reshape([256,256,256])

temp = np.genfromtxt(temp_dir)
temp = temp.reshape([256,256,256])
```

```python
info_list = np.array(list(zip(data_num,data_radius_grid,data_coordx,data_coordy,data_coordz)))
```

```python
xray_temp =[]
for nn,(num,r,ix,iy,iz) in enumerate(info_list):
    #print(nn,num,int(round(r/0.39,2)),int(ix),int(iy),int(iz))
    ix = int(ix)
    iy = int(iy)
    iz = int(iz)
    rn = int(r)
    
    
    xray_sum = np.sum(xray[ix-rn:ix+rn+1,iy-rn:iy+rn+1,iz-rn:iz+rn+1])
    xray_temp_sum = np.sum(np.dot ( temp[ix-rn:ix+rn+1,iy-rn:iy+rn+1,iz-rn:iz+rn+1] , xray[ix-rn:ix+rn+1,iy-rn:iy+rn+1,iz-rn:iz+rn+1]))
    
    xray_temp.append((xray_temp_sum/xray_sum)/1.16*10**7)
```

```python
df = pd.DataFrame({'number':data_num,'x':data_coordx,'y':data_coordy,'z':data_coordz,'virial_radius (Mpc)':data_radius, \
                  'data_mass (log)': data_log_mass,'xray_weighted_temp(Kev)'xray_temp})
df.to_excel('/storage/cluster/data/data.xlsx')
```
