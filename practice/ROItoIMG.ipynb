```python
from data_cleaning_util import *
import os
import cv2
import matplotlib.pyplot as plt

#roi_path는 roi 압축파일들을 포함하는 폴더
roi_path = 'G:\\JBS-05K\\2차\\annotation'
```

```python
save_path = 'G:\\JBS-05K\\2차\\ROI'
if not os.path.isdir(save_path):
    os.mkdir(save_path)
for name in os.listdir(roi_path):
    print(name[:-4])
    r = get_roi_arr(roi_path,name)
    ret,pos = roizip2fig('',roi_path,name,isnii=True)
    ret = np.array(ret)
    if not os.path.isdir(os.path.join(save_path,name[:-4])):
        os.mkdir(os.path.join(save_path,name[:-4]))
    for i in np.unique(pos):
        roi = np.max(ret[pos == i],axis=0)
        
        plt.imsave(os.path.join(*[save_path,name[:-4],"slice-" + str(i) + ".jpg"]), roi)
        #cv2.imwrite(os.path.join(*[save_path,name[:-4],"slice-" + str(i) + ".jpg"]), roi)
```
